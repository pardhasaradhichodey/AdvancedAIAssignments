```python
#The following variables must be used in your program. You can change the location values for testing purposes
stoneLocation=[2, 8]  #[row, column]
positiveTerminalLocation = [0,9,2]  #[row, column, reward value]
negativeTerminalLocation = [1,9, -2]  #[row, column, reward value]
iteration=30
noise=0.15
discount=0.91
```

```python
def get_max_var(var1, var2, var3, var4):
    max_var = None
    max_val = float('-inf')
    for var_name, var_val in locals().items():
        if var_name.startswith('var') and var_val > max_val:
            max_var = var_name
            max_val = var_val
    return max_var, max_val
def iteration_fun(grid,dir_grid):
    final_grid=[[0 for i in range(10)] for j in range(10)]
    final_dir_grid=[['n' for i in range(10)] for j in range(10)]
    north=south=east=west=0
    for i in range(10):
        for j in range(10):
            if(i==stoneLocation[0] and j==stoneLocation[1]):
                final_grid[i][j]=None
                final_dir_grid[i][j]='STONE'
                continue
            if i==positiveTerminalLocation[0] and j==positiveTerminalLocation[1]:
                final_grid[i][j]=positiveTerminalLocation[2]
                final_dir_grid[i][j]='n'
                continue
            if i==negativeTerminalLocation[0] and j==negativeTerminalLocation[1]:
                final_grid[i][j]=negativeTerminalLocation[2]
                final_dir_grid[i][j]='n'
                continue
            if i==0:
                north=0
            elif i-1==stoneLocation[0] and j==stoneLocation[1]:
                north=0
            else:
                north=grid[i-1][j]*discount
            if i==9:
                south=0
            elif i+1==stoneLocation[0] and j==stoneLocation[1]:
                south=0
            else:
                south=grid[i+1][j]*discount
            if j==9:
                east=0
            elif i==stoneLocation[0] and j+1==stoneLocation[1]:
                east=0
            else:
                east=grid[i][j+1]*discount
            if j==0:
                west=0
            elif i==stoneLocation[0] and j-1==stoneLocation[1]:
                west=0
            else:
                west=grid[i][j-1]*discount
            
            temp_noise=noise/2
            f_north=north*(1-noise)+(east+west)*temp_noise
            f_south=south*(1-noise)+(east+west)*temp_noise
            f_east=east*(1-noise)+(south+north)*temp_noise
            f_west=west*(1-noise)+(south+north)*temp_noise
            var1,var2,var3,var4=f_north,f_south,f_east,f_west
            max_var, max_val = get_max_var(var1, var2, var3, var4)
            final_grid[i][j]=max_val
            if max_var=='var1':
                final_dir_grid[i][j]='n'
            elif max_var=='var2':
                final_dir_grid[i][j]='s'
            elif max_var=='var3':
                final_dir_grid[i][j]='e'
            else:
                final_dir_grid[i][j]='w'
    return final_grid,final_dir_grid
```

```python
def print_grid(grid,dir_grid):
    for i in range(10):
        for j in range(10):
            if j==9:
                if grid[i][j]==None:
                    print('(STONE)')
                else:
                    #print(f"{grid[i][j]}({dir_grid[i][j]})",end=',') #If you want print the entire values of each position put the next line in comment and remove this line from comments 
                    print("%.2f" %grid[i][j],f"({dir_grid[i][j]})")
            else:
                if grid[i][j]==None:
                    print('(STONE)',end=',')
                else:
                    #print(f"{grid[i][j]}({dir_grid[i][j]})",end=',') #If you want print the entire values of each position put the next line in comment and remove this line from comments 
                    print("%.2f" %grid[i][j],f"({dir_grid[i][j]})",end=',')
```

```python
def start():
    grid=[[0 for i in range(10)] for j in range(10)]
    dir_grid=[['n' for i in range(10)] for j in range(10)]
    grid[positiveTerminalLocation[0]][positiveTerminalLocation[1]]=positiveTerminalLocation[2]
    grid[negativeTerminalLocation[0]][negativeTerminalLocation[1]]=negativeTerminalLocation[2]
    grid[stoneLocation[0]][stoneLocation[1]]=None
    dir_grid[stoneLocation[0]][stoneLocation[1]]='STONE'
    print('Iteration Output: 1')
    print_grid(grid,dir_grid)
    for i in range(iteration-1):
        print(f"Iteration Output: {i+2}")
        grid,dir_grid=iteration_fun(grid,dir_grid)
        print_grid(grid,dir_grid)
```
